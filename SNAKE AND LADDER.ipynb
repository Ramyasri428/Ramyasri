```python
import random
ladder = {(3,11),(14,22),(18,34),(26,54),(47,77),(61,80),(65,90),(73,87),(88,95),}
snake = {(18,2),(27,16),(39,23),(40,8),(49,27),(56,6),(70,29),(85,28),(98,6)}
position1 = 0
position2 = 0
```

```python
def move(position):
    dice = random.randint(1,6)
    print(f"dice:{dice}")
    position= position+dice
    
    if position in snake :
        print("snake bitten")
        position = snake[position]
        print(f"position: {position}")
        
    elif position in ladder :
        print("climb throw ladder")
        position = ladder[position]
        print(f"position: {position}")
        
    else:
        print(f"position: {position}")
    print("\n")     
    return position
    
while True:
        A = input('player1 "A"  Throw a dice: ')
        position1 = move(position1)
        if position1 >= 100:
            print("Game over, Player1 won the game")
            break
              
        B = input('player2 "B"   Throw a dice')
        position2 = move(position2)
        if position2 >= 100:
            print("Game over ,Player2 won the game")      
            break
```
